```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("car_prices.csv")
```

```python
data = df.copy()
```

```python
data
```

```python
data = data.dropna()
```

```python
data
```

```python
data.describe()
```

```python
cars_2010 = data.loc[lambda table: table["year"] == 2010]
```

```python
cars_2010
```

```python
car_make_counts = cars_2010["make"].value_counts()
```

```python
top_cars_2010 = dict(car_make_counts.head())
```

```python
top_cars_2010
```

```python
top_cars_2010.keys()
```

```python
sns.set()
f, ax = plt.subplots(figsize=(20, 5))

bars = sns.barplot(x=list(top_cars_2010.keys()), y=list(top_cars_2010.values()), ax=ax)

ax.set_title("Top 5 Car Brands Sold in 2010")
ax.set_xlabel("Brands")
ax.set_ylabel("Cars Sold")
ax.set_ylim((0, 5000))

for bar in bars.patches:
    ax.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), 
            int(bar.get_height()), ha='center', va='bottom')

plt.show()
```

```python
toyota_cars = data.loc[(data["make"] == "Ford") & (data["year"] == 2010)]
```

```python
toyota_cars
```

```python
toyota_cars_counts = toyota_cars["body"].value_counts()
```

```python
toyota_cars_counts 
```

```python
top_ford_2010 = dict(toyota_cars_counts.head())
```

```python
top_ford_2010
```

```python
sns.set()
f, ax = plt.subplots(figsize=(20, 5))

bars = sns.barplot(x=list(top_ford_2010.keys()), y=list(top_ford_2010.values()), ax=ax)

ax.set_title("Top 5 Car Brands Sold in 2010")
ax.set_xlabel("Body")
ax.set_ylabel("Cars Sold")
ax.set_ylim((0, 1000))

for bar in bars.patches:
    ax.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), 
            int(bar.get_height()), ha='center', va='bottom')

plt.show()
```
